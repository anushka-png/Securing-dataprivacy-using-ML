```python
df = pd.read_csv(r"C:\Users\DELL\Desktop\semesters\SYSTEM_FILE.csv")
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
bow_vectorizer = CountVectorizer(max_df=0.90,min_df=2,max_features=1000)
bow = bow_vectorizer.fit_transform(df['detected as User/attacker'])
```

```python
bow[0].toarray()
```

```python
from sklearn.model_selection import train_test_split
X_train, x_test,y_train,y_test = train_test_split(bow,df['action required'],random_state=100,test_size=0.25)
```

```python

#  The F1 score is defined as the harmonic mean of precision and recall.
## precision --->  Within everything that has been predicted as a positive, precision counts the percentage that is correct:
### recall --> Within everything that actually is positive, how many did the model succeed to find

#### F1_SCORE =2PR/P+R ########

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score,accuracy_score
```

```python
##Logistic regression becomes a classification technique only when a decision threshold is brought into the picture.(attacker or user


model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
pred = model.predict(x_test)
f1_score(y_test,pred)
```

```python
# by f1_score we get  to what extent the model is capable of  showing correct result. by combining the metrics of both precision and recall metircs into a single one ./
```

```python
accuracy_score(y_test,pred)   
```

```python
## our model is performing correctly with an  accuracy of 100 percent 
```
